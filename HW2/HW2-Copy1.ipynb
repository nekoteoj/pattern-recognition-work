```python
def histogram(df: pd.DataFrame, col: str, bin_size: int):
    # remove NaN values
    df_col_no_nan = df[~np.isnan(df[col])][col]
    # hist is the count for each bin
    # bin_edge is the edge values of the bins
    hist, bin_edge = np.histogram(df_col_no_nan, bin_size)
    return hist, bin_edge
```

```python
bin_zero_count = 0
def plot_histogram(df: pd.DataFrame, col: str, bin_size: int):
    hist, bin_edge = histogram(df, col, bin_size)
    global bin_zero_count
    bin_zero_count += len(hist[hist == 0])
    print(hist)
    # plot the histogram
    plt.fill_between(bin_edge.repeat(2)[1:-1],hist.repeat(2),facecolor='steelblue')
    plt.title(col)
    plt.show()
```

```python
bin_edges = {}
hists = {}
def bin_column(df: pd.DataFrame, col: str, bin_size: int = 10):
    hist, bin_edge = histogram(df, col, bin_size)
    global bin_edges
    global hists
    bin_edge = bin_edge[:-1]
    bin_edges[col] = bin_edge
    hists[col] = hist
    df.loc[~np.isnan(df[col]), col] = np.digitize(df[~np.isnan(df[col])][col], bin_edge)
```

```python
saved_MLE = {}
def get_MLE(x, col, train):
    global saved_MLE
    if col in saved_MLE:
        if x in saved_MLE[col]:
            return saved_MLE[col][x]
    else:
        saved_MLE[col] = {}
    col_w1 = train[(~np.isnan(train['Attrition'])) & (train['Attrition'] == 1.0)][col]
    col_w2 = train[(~np.isnan(train['Attrition'])) & (train['Attrition'] == 0.0)][col]
    MLE_w1 = len(col_w1[col_w1 == x]) / len(col_w1)
    MLE_w2 = len(col_w2[col_w2 == x]) / len(col_w2)
    if MLE_w1 == 0.0:
        MLE_w1 = 1e-20
    if MLE_w2 == 0.0:
        MLE_w2 = 1e-20
    saved_MLE[col][x] = (MLE_w1, MLE_w2)
    return (MLE_w1, MLE_w2)
```

```python
def diff_log_MLE(x, col, train):
    MLE_w1, MLE_w2 = get_MLE(x, col, train)
    return np.log(MLE_w1) - np.log(MLE_w2)
```

```python
def get_prior(train_y):
    p_leave = train_y[train_y == 1.0].shape[0] / train_y.shape[0]
    p_stay = train_y[train_y == 0.0].shape[0] / train_y.shape[0]
    return (p_leave, p_stay)
```

```python
def which_bin(bin_edge, x):
    bin_edge = bin_edge.copy()
    if x.min() < bin_edge[0]:
        bin_edge[0] = x.min()
    if x.max() > bin_edge[-1]:
        bin_edge[-1] = x.max()
    return np.digitize(x, bin_edge)
```

```python
def report(predicted, test):
    true_pos = len(predicted[(predicted == 1.0) & (test['Attrition'] == 1.0)])
    true_neg = len(predicted[(predicted == 0.0) & (test['Attrition'] == 0.0)])
    false_pos = len(predicted[(predicted == 1.0) & (test['Attrition'] == 0.0)])
    false_neg = len(predicted[(predicted == 0.0) & (test['Attrition'] == 1.0)])
    accuracy = (true_pos + true_neg) / len(predicted)
    print(true_pos, true_neg, false_pos, false_neg)
    if true_pos > 0:
        recall = true_pos / (true_pos + false_neg)
        precision = true_pos / (true_pos + false_pos)
        f1 = 2 / (1/recall + 1/precision)
    else:
        recall = 0.0
        precision = 0.0
        f1 = 0.0
    print('Accuracy:', accuracy)
    print('Recall:', recall)
    print('Precision:', precision)
    print('F1:', f1)
    return accuracy, f1
```

```python
def naive_bayes_classifier(train: pd.DataFrame, test: pd.DataFrame, threshold = 0.0):
    # Prepare Data
    global saved_MLE
    train = train.copy()
    train_y = train['Attrition']
    train_x = train.drop(columns=['Attrition'])
    
    test_x = test.drop(columns=['Attrition']).copy()
    
    p_leave, p_stay = get_prior(train_y)
    
    for col in tqdm(test_x.columns):
        if col in bin_edges:
            test_x.loc[~np.isnan(test_x[col]), col] = which_bin(bin_edges[col], test_x[~np.isnan(test_x[col])][col].values)
        test_x.loc[~np.isnan(test_x[col]), col] = test_x[~np.isnan(test_x[col])][col].apply(diff_log_MLE, args=(col, train))
        test_x.loc[np.isnan(test_x[col]), col] = 0.0
    
    test_x['Predicted'] = test_x.sum(axis=1)
    test_x['Predicted'] += np.log(p_leave) - np.log(p_stay)
    
    predicted = test_x['Predicted'].values
    predicted[predicted > threshold] = 1.0
    predicted[predicted <= threshold] = 0.0
    report(predicted, test)
    return predicted
```

```python
def naive_bayes_gaussian_classifier(train: pd.DataFrame, test: pd.DataFrame, threshold = 0.0):
    train_w1 = train[train['Attrition'] == 1.0]
    train_w2 = train[train['Attrition'] == 0.0]
    
    train_w1_mean = train_w1.mean()
    train_w2_mean = train_w2.mean()
    
    train_w1_std = train_w1.std()
    train_w2_std = train_w2.std()
    
    train_w1_std[train_w1_std == 0.0] = 1e-15
    train_w2_std[train_w2_std == 0.0] = 1e-15
    
    p_leave, p_stay = get_prior(train['Attrition'])
    
    test_x = test.drop(columns=['Attrition']).copy()
    
    for col in test_x:
        l_w1 = scipy.stats.norm(train_w1_mean[col], train_w1_std[col]).pdf(test_x[~np.isnan(test_x[col])][col])
        l_w2 = scipy.stats.norm(train_w2_mean[col], train_w2_std[col]).pdf(test_x[~np.isnan(test_x[col])][col])
        test_x.loc[~np.isnan(test_x[col]), col] = np.log(l_w1) - np.log(l_w2)
        test_x.loc[np.isnan(test_x[col]), col] = 0.0
    
    test_x['Predicted'] = test_x.values.sum(axis=1)
    test_x['Predicted'] += np.log(p_leave) - np.log(p_stay)
    test_x['Predicted'] = test_x['Predicted'] > threshold
    test_x.loc[test_x['Predicted'] == True, 'Predicted'] = 1.0
    test_x.loc[test_x['Predicted'] == False, 'Predicted'] = 0.0
    
    predicted = test_x['Predicted'].values
    

    report(predicted, test)
    return predicted
```

```python
def TPR_FPR(predicted, test):
    true_pos = len(predicted[(predicted == 1.0) & (test['Attrition'] == 1.0)])
    true_neg = len(predicted[(predicted == 0.0) & (test['Attrition'] == 0.0)])
    false_pos = len(predicted[(predicted == 1.0) & (test['Attrition'] == 0.0)])
    false_neg = len(predicted[(predicted == 0.0) & (test['Attrition'] == 1.0)])
    if true_pos > 0:
        tpr = true_pos / (true_pos + false_neg)
    else:
        tpr = 0.0
    if false_pos > 0:
        fpr = false_pos / (false_pos + true_neg)
    else:
        fpr = 0.0
    return fpr, tpr
```
